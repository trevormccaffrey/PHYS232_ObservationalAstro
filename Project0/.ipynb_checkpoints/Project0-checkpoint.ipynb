Only take a circular region centered on the image.  This is basically just a problem of formatting a "circular" mask in a 2D array:

```python
#see https://www.semicolonworld.com/question/44279/how-to-apply-a-disc-shaped-mask-to-a-numpy-array
def circle_mask(data, R, center=None):
    if center is None: xc, yc = data.shape[0]//2, data.shape[1]//2
    y,x = np.ogrid[0: data.shape[1], 0: data.shape[0]]
    mask = (x-xc)**2+(y-yc)**2 <= R**2
    return np.ma.array(data, mask=~mask)
```

## The Importance of Cooling
How does the performance of the CCD vary with temperature?  Check bias and dark images at different temperatures.

```python
log.head()
```

```python
obs_type = log["Object"].values
seq_num  = log["Seq. Num."].values
flag     = log["Flag"].values
temps    = log["Temp (C)"].values
times    = log["Exp (s)"].values
prefix   = "2021Oct08/Project0_TrevorAdamKen"

#The counts and temps for darks also depend on exposure time, so save them as dictionaries
mean_dark = {
    "1": np.array([]),
    "5": np.array([]),
    "20": np.array([]),
    "60": np.array([])
}
temp_dark = {
    "1": np.array([]),
    "5": np.array([]),
    "20": np.array([]),
    "60": np.array([])
}
mean_bias   = []
temp_bias   = []

R = 350 #radius of region to measure mean counts, unit is pixels

for i in range(log.shape[0]):
    if "-" in seq_num[i]:
        ind_min, ind_max = np.array(seq_num[i].split("-")).astype(int)
        for j in range(ind_min, ind_max+1):
            image = fits.getdata(prefix + obs_type[i].capitalize() + ".%08d.fit"%j)
            if obs_type[i].lower()=="dark" and flag[i]:
                mean_dark["%d"%times[i]] = np.append(mean_dark["%d"%times[i]], np.mean(circle_mask(image, R))) #take mean of circular region at center
                temp_dark["%d"%times[i]] = np.append(temp_dark["%d"%times[i]], temps[i])
            elif obs_type[i].lower()=="bias" and flag[i]:
                mean_bias.append(np.mean(circle_mask(image, R)))
                temp_bias.append(temps[i])

    else:
        image = fit.getdata(prefix + obs_type[i].capitalize() + ".%08d.fit"%seq_num[i])
        if obs_type[i].lower()=="dark" and flag[i]:
            mean_dark["%d"%times[i]] = np.append(mean_dark["%d"%times[i]], np.mean(circle_mask(image, R))) #take mean of circular region at center
            temp_dark["%d"%times[i]] = np.append(temp_dark["%d"%times[i]], temps[i])
        elif obs_type[i].lower()=="bias" and flag[i]:
            mean_bias.append(np.mean(circle_mask(image, R)))
            temp_bias.append(temps[i])
        
#mean_dark = np.array(mean_dark)
#temp_dark = np.array(temp_dark)
mean_bias = np.array(mean_bias)
temp_bias = np.array(temp_bias)
```

Errors are Poisson, so can get them as $\sqrt{N}$

```python
errs_dark = {
    "1": np.sqrt(mean_dark["1"]),
    "5": np.sqrt(mean_dark["5"]),
    "20": np.sqrt(mean_dark["20"]),
    "60": np.sqrt(mean_dark["60"])
}
errs_bias = np.sqrt(mean_bias)
```

```python
fig = plt.figure(figsize=(6.5,6.5))
plt.errorbar(temp_dark["1"], mean_dark["1"], yerr=errs_dark["1"], fmt="o", label="Exp. Time = 1s")
plt.errorbar(temp_dark["5"], mean_dark["5"], yerr=errs_dark["5"], fmt="o", label="Exp. Time = 5s")
plt.errorbar(temp_dark["20"], mean_dark["20"], yerr=errs_dark["20"], fmt="o", label="Exp. Time = 20s")
plt.errorbar(temp_dark["60"], mean_dark["60"], yerr=errs_dark["60"], fmt="o", label="Exp. Time = 60s")
plt.xlabel("Temperature (Celsius)", fontsize=20)
plt.xticks(fontsize=15)
plt.ylabel("Counts", fontsize=20)
plt.yticks(fontsize=15)
plt.legend(loc="best", prop={"size":17})
plt.title("Dark Images", fontsize=20)
plt.show()
```

The counts in the dark images increase with temperature, and also strongly depend on total exposure time: as the detector is exposed for a longer amount of time, the amount of charge that builds goes up.

You need to go to at least 30deg Celsius below ambient ($\sim-10^{\circ}\mathrm{C}$ in plot) to reach consistent performance for different exposure times.  Going colder at first may make results even more consistent, but we'd eventually start to build up frost on the CCD, which would end up *degrading* the performance quality of the CCD.

```python
fig = plt.figure(figsize=(6.5,6.5))
plt.errorbar(temp_bias, mean_bias, yerr=errs_bias, fmt="o")
plt.xlabel("Temperature (Celsius)", fontsize=20)
plt.xticks(fontsize=15)
plt.ylabel("Counts", fontsize=20)
plt.yticks(fontsize=15)
plt.title("Bias Images", fontsize=20)
plt.show()
```

The counts in bias images go up with Temperature.

From our data, we found that Bias frames held more counts when the CCD was warmer. It appears that the relationship between counts decreases exponentially, with the number of counts varying far less as the CCD reached lower temperatures.

Dark frames also experienced similar exponential decreases in counts as temperature decreased. Our data also indicates a large increase in counts in long exposure frames while at high temperatures, where lower temperatures had very little effect on the long exposure dark frames.

Our data show that we had the most consistent results at temperatures of 20-30 degrees below ambient, with slightly more consistent dark frame results closer to 30 degrees below ambient.
Colder temperatures would help, but our data indicates that the change in counts is significantly smaller as the CCD gets colder and colder. There is also the chance of freezing any moisture that could be inside the CCD.

# Linearity
This time we'll evaluate how the mean counts vary as a function of exposure time -- leaving temperature constant (~30deg Celsius below ambient).  The CCD should have a mostly linear response, but become inconsistent as it starts to become more saturated.

```python
!ls
```

The second log is the same format as first, except is for the linearity section.  Has different exposure times and counts, more flat field images.

```python
log2 = pd.read_csv("DLOlog8Oct2021_TrevorAdamKen_Project0_Linearity.csv")
log2.head()
```

```python
obs_type = log2["Object"].values
seq_num  = log2["Seq. Num."].values
flag     = log2["Flag"].values
temps    = log2["Temp (C)"].values
times    = log2["Exp (s)"].values
prefix   = "2021Oct08/Project0_TrevorAdamKen"

#The counts and temps for darks also depend on exposure time, so save them as dictionaries
mean_counts   = []
exp_time      = []

R = 350 #radius of region to measure mean counts, unit is pixels

for i in range(log2.shape[0]):
    if obs_type[i].lower()=="flat" and flag[i]: #only look a non-saturated flats
        if "-" in seq_num[i]:
            ind_min, ind_max = np.array(seq_num[i].split("-")).astype(int)
            for j in range(ind_min, ind_max+1):
                image = fits.getdata(prefix + "FlatFieldBlue.%08d.fit"%j)
                mean_counts.append(np.mean(circle_mask(image, R))) #take mean counts of pixels within radius R from center
                exp_time.append(times[i])

        else:
            image = fits.getdata(prefix + "FlatFieldBlue.%08d.fit"%int(seq_num[i]))
            mean_counts.append(np.mean(circle_mask(image, R))) #take mean counts of pixels within radius R from center
            exp_time.append(times[i])
        
#mean_dark = np.array(mean_dark)
#temp_dark = np.array(temp_dark)
mean_counts = np.array(mean_counts)
exp_time    = np.array(exp_time)
```

```python
#Again, 1sigma errors are Poisson
errs_counts = np.sqrt(mean_counts)
```

```python
errs_counts
```

```python
mean_counts
```

```python
#Determine a linear fit between ~5000-45000 counts
good_mask = ((mean_counts>5000) & (mean_counts<45000))
m, b = np.polyfit(exp_time[good_mask], mean_counts[good_mask], 1) #determine fit parameters only in linear region of CCD
x = np.linspace(min(exp_time), max(exp_time), 10) #fit on region, but display over whole range
y = m*x + b

fig = plt.figure(figsize=(6.5,6.5))
plt.errorbar(exp_time, mean_counts, yerr=errs_counts, fmt="o")
plt.plot(x, y, "--k")
plt.xlabel("Exposure Time (s)", fontsize=20)
plt.ylabel("Counts", fontsize=20)
plt.xticks(fontsize=15)
plt.yticks(fontsize=15)

plt.text(10, 5000, "y = %.2f*x + %.2f"%(m,b), fontsize=22.5)

plt.show()
```

The CCD response is generally linear up until ~40 seconds of exposure time.  Even in the region where it is linear, we found the count measurements to sometimes be fairly scattered for the same exposure times.

# Gain and Read Noise
Now we'll finally determine the gain and read noise of the CCD.  We'll need to subtract a mean "superbias" and "superdark" from each of our flat field images.

1) The "superbias" will be a single image.  Each pixel will be the mean of the pixel values at that location for all bias images.  We'll subtract this from *every* other image.

2) The "superdarks" will be multiple images: there will be one for each flat field with the same exposure time.  We'll take the mean of each pixel as above.  These will be subtracted from the flat field with identical exposure time.

3) We'll also take "difference frames" for each pair of bias and flat image with the same exposure time.  This will be computed as flat$_{i_1}-$flat$_{i_2}$ **(but these won't go into any calculations???)**

---


First load in all "good" bias frames, then mean stack them to create the "super bias" image:

```python
#Create the "superbias" first
obs_type = log2["Object"].values
seq_num  = log2["Seq. Num."].values
flag     = log2["Flag"].values
temps    = log2["Temp (C)"].values
times    = log2["Exp (s)"].values
prefix   = "2021Oct08/Project0_TrevorAdamKen"

#Load all bias images in list
images_bias = []
for i in range(log2.shape[0]):
    if obs_type[i].lower()=="bias" and flag[i]:
        if "-" in seq_num[i]:
            ind_min, ind_max = np.array(seq_num[i].split("-")).astype(int)
            for j in range(ind_min, ind_max+1):
                image = fits.getdata(prefix + "Bias.%08d.fit"%j)
                images_bias.append(image)

        else:
            image = fit.getdata(prefix + "Bias.%08d.fit"%int(seq_num[i]))
            images_bias.append(image)

images_bias = np.array(images_bias)
```

```python
super_bias = np.mean(images_bias, axis=0)
plt.imshow(super_bias, cmap="gray", interpolation="nearest", vmin=np.percentile(super_bias, 5), vmax=np.percentile(super_bias, 95))
cbar = plt.colorbar()
cbar.ax.set_ylabel("Counts")
plt.xticks([])
plt.yticks([])
plt.title("Super Bias", fontsize=20)
plt.show()
```

Now make super darks -- one for each exposure time.

```python
np.unique(times[flag==1])
```

```python
images_dark = {}
# initialize the images with our given exposure times
for time in np.unique(times[flag==1])[1:]: #ignore 0s
    images_dark["%.1f"%time] = []

for i in range(log2.shape[0]):
    if obs_type[i].lower()=="dark" and flag[i]:
        if "-" in seq_num[i]:
            ind_min, ind_max = np.array(seq_num[i].split("-")).astype(int)
            for j in range(ind_min, ind_max+1):
                image = fits.getdata(prefix + "Dark.%08d.fit"%j) - super_bias #Change this for darks?? TM: but I think we need it?
                images_dark["%.1f"%times[i]].append(image)

        else:
            image = fit.getdata(prefix + "Dark.%08d.fit"%int(seq_num[i])) - super_bias
            images_dark["%.1f"%times[i]].append(image)
```

```python
#images_dark has all darks organized by exposure time; stack them to get a "super dark" for each exposure time
super_darks = {
    "2.5" : np.mean(np.array(images_dark["2.5"]), axis=0),
    "5.0" : np.mean(np.array(images_dark["5.0"]), axis=0),
    "8.0" : np.mean(np.array(images_dark["8.0"]), axis=0),
    "12.0": np.mean(np.array(images_dark["12.0"]), axis=0),
    "18.0": np.mean(np.array(images_dark["18.0"]), axis=0),
    "25.0": np.mean(np.array(images_dark["25.0"]), axis=0),
    "32.5": np.mean(np.array(images_dark["32.5"]), axis=0),
    "40.0": np.mean(np.array(images_dark["40.0"]), axis=0),
    "42.0": np.mean(np.array(images_dark["42.0"]), axis=0),
    "43.0": np.mean(np.array(images_dark["43.0"]), axis=0),
    "45.0": np.mean(np.array(images_dark["45.0"]), axis=0),
    "47.5": np.mean(np.array(images_dark["47.5"]), axis=0),
    "50.0": np.mean(np.array(images_dark["50.0"]), axis=0)
}
```

And now the flat-field images; subtract the appropriate super dark and the super bias from each.

```python
images_flat = {}
# initialize the images with our given exposure times
for time in np.unique(times[flag==1])[1:]: #can skip 0s
    images_flat["%.1f"%time] = []

for i in range(log2.shape[0]):
    if obs_type[i].lower()=="flat" and flag[i]:
        if "-" in seq_num[i]:
            ind_min, ind_max = np.array(seq_num[i].split("-")).astype(int)
            for j in range(ind_min, ind_max+1):
                image = fits.getdata(prefix + "FlatFieldBlue.%08d.fit"%j) - super_bias - super_darks["%.1f"%times[i]] #subtract averaged bias and dark frames
                images_flat["%.1f"%times[i]].append(image)

        else:
            image = fits.getdata(prefix + "FlatFieldBlue.%08d.fit"%int(seq_num[i])) - super_bias - super_darks["%.1f"%times[i]]
            images_flat["%.1f"%times[i]].append(image)
```

Now get variance and mean of flat-field images.

```python
list(images_flat.keys())
```

```python
mean_flat = [] #the mean counts of each flat
std_flat  = [] #the standard deviation of each flat image

for key in list(images_flat.keys()):
    for img in images_flat[key]:
        std_flat.append(np.std(img))
        mean_flat.append(np.mean(img))

mean_flat = np.array(mean_flat)
std_flat = np.array(std_flat)
```

```python
good_mask = (mean_flat>30000)
m, b = np.polyfit(mean_flat[good_mask], std_flat[good_mask]**2, 1)
x = np.linspace(22000, 56000)
y = m*x + b

fig = plt.figure(figsize=(7,7))
plt.scatter(mean_flat, std_flat**2)
plt.plot(x, y, "--k", alpha=0.7)
plt.xlabel("Mean Counts", fontsize=20)
plt.ylabel("$\sigma^2$ (variance of counts)", fontsize=20)
plt.xticks(fontsize=15)
plt.yticks(fontsize=15)
plt.text(500, 1.5e6, "y = %.2f*x + %.2f"%(m,b), fontsize=22.5)
plt.show()
```

Using the equation:

$$ \sigma^2 = \frac{N}{g} + \frac{\sigma_\mathrm{R}^2}{g^2} $$ 

We recognize that the gain of the detector is just the inverse of linear slope of the photon noise regime in the above plot, **1/45 = 0.02 electrons/count**.

```python
m, b = np.polyfit(np.log10(mean_flat), np.log10(std_flat), 1)
x = np.linspace(3.3, 4.9, 10)
y = m*x + b

fig = plt.figure(figsize=(7,7))
plt.scatter(np.log10(mean_flat), np.log10(std_flat))
plt.plot(x, y, "--k", alpha=0.7)
plt.ylim(1,None)
plt.ylabel("$\log{\sigma}$", fontsize=20)
plt.xlabel("$\log{N}$", fontsize=20)
plt.xticks(fontsize=15)
plt.yticks(fontsize=15)
plt.text(3.5, 1.25, "y = %.2f*x + %.2f"%(m,b), fontsize=22.5)
plt.show()
```


With a y-intercept of -1.4, **we find a read noise of $10^{-1.4} = 0.04 $ counts** or 2 electrons per count.  We note that this is likely an underestimate as our plot never truly appears to flatten out to the "low noise regime".
